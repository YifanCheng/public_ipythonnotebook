```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
raw = pd.read_csv('../simple_test_case/output/simple_test/36.0625_-83.6875/36.0625_-83.6875_reach1_seg1', sep=' ', header=None,parse_dates=[[0,1,2]])
```

```python
raw = raw.rename(columns={'0_1_2': 'time', 3: 'flow(cfs)', 4: 'streamT', 5: 'headwaterT', 6: 'airT'})
raw = raw.set_index('time')
```

```python
timeperiod = pd.date_range('1949-01-01',periods=500)
```

# This is for outlet of the simple test case

```python

plt.figure(figsize=[10,7.2])

f1, = plt.plot(timeperiod,raw['streamT'].loc[timeperiod], label = 'Outflow Temperature')

f2 , =plt.plot(timeperiod,raw['headwaterT'].loc[timeperiod], label = 'Headwater Temperature')

plt.legend(handles = [f1,f2], loc = 0)
plt.title('Outflow temperature versus Headwater temperature in simple test case', fontsize = 16)
plt.xlabel('Date', fontsize = 14)
plt.ylabel('Temperature(' + r"$^o$" + 'C)', fontsize = 14)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)

plt.savefig('./outflowT-headwaterT.png', dpi = 200, bbox_inches = 'tight')
```
